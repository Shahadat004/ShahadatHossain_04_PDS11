```python
# Project 2: Dataset Merging, Data Manipulation, K-Means Clustering


# Task P1: Create CSV file for Term Test 1 Results
import random
import pandas as pd
files1 = {
    'Registration Number': [f'202500{i:02d}' for i in range(1, 51)],
     'Name': [f'Student_{i}' for i in range(1, 51)],
     'TT-1 Marks': [random.randint(5, 20) for _ in range(50)]
    }
df1 = pd.DataFrame(files1)

files2 = {
    'Registration Number': [f'202500{i:02d}' for i in range(1, 51)],
     'Name': [f'Student_{i}' for i in range(1, 51)],
     'TT-2 Marks': [random.randint(5, 20) for _ in range(50)]
    }
df2 = pd.DataFrame(files2)
```

```python
# Task P1: Create and Merge Term Test Results

merged_df = pd.merge(df1, df2, on=['Registration Number', 'Name'])

merged_df
```

```python
# 4.

# Create a new column for the best term test marks (higher of TT-1 and TT-2 Marks)
merged_df['Best Term Test Marks'] = merged_df[['TT-1 Marks', 'TT-2 Marks']].max(axis=1)

# Create a new column for the average term test marks (average of TT-1 and TT-2 Marks)
merged_df['Average Term Test Marks'] = merged_df[['TT-1 Marks', 'TT-2 Marks']].mean(axis=1)

# Display the updated DataFrame
print(merged_df.head())
```

```python
# 5.

# Drop the 'TT-1 Marks' and 'TT-2 Marks' columns
merged_df = merged_df.drop(columns=['TT-1 Marks', 'TT-2 Marks'])

# Display the updated DataFrame
print(merged_df.head())
```

```python
# 6.

import pandas as pd
import random

# Step 1: Create the attendance and term final marks CSV file
attendance_data = {
    'Registration Number': [f'202500{i:02d}' for i in range(1, 51)],  # Same Registration Numbers
    'Name': [f'Student_{i}' for i in range(1, 51)],  # Include the 'Name' column
    'Attendance Marks': [random.randint(8, 10) for _ in range(50)],  # Random attendance marks between 8 and 10
    'Term Final Marks': [random.randint(50, 100) for _ in range(50)]  # Random term final marks between 70 and 100
}

# Create a DataFrame for attendance and term final marks
attendance_df = pd.DataFrame(attendance_data)

# Save the DataFrame to a CSV file
attendance_df.to_csv('attendance_term_final_marks.csv', index=False)


# Load the attendance and term final marks data
attendance_df = pd.read_csv('attendance_term_final_marks.csv')
attendance_df
```

```python
# Convert 'Registration Number' to a consistent data type (e.g., string) in both DataFrames before merging.
merged_df['Registration Number'] = merged_df['Registration Number'].astype(str)
attendance_df['Registration Number'] = attendance_df['Registration Number'].astype(str)

final_df= pd.merge(merged_df, attendance_df, on=['Registration Number', 'Name'])

final_df['Final Marks'] = (final_df['Term Final Marks'] * 0.7) + final_df['Average Term Test Marks'] + final_df['Attendance Marks']

print(final_df.head())
```

```python
# 7.

# Write the content of the final dataframe to a new CSV file
final_df.to_csv('final result.csv', index=False)

# Display a message to confirm the file has been saved
print("The 'final result.csv' file has been saved successfully!")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Assuming final_df is already defined with a 'Final Marks' column
X = final_df[['Final Marks']]

# Standardizing the data (optional but helps with clustering performance)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Finding the optimal number of clusters using the Elbow method
inertia = []
K_range = range(2, 6)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

# Plot the Elbow method graph
plt.figure(figsize=(8, 5))
plt.plot(K_range, inertia, marker='o', linestyle='--')
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia")
plt.title("Elbow Method for Optimal K")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans

# Assuming 'final_df' is your DataFrame and contains 'Final Marks'
X = final_df[['Final Marks']]

# Apply K-Means Clustering
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
final_df['Cluster'] = kmeans.fit_predict(X)

# Reorder clusters based on mean marks
cluster_order = final_df.groupby('Cluster')['Final Marks'].mean().sort_values().index

# Create a mapping: lowest avg mark → Cluster 0, medium → Cluster 1, highest → Cluster 2
cluster_mapping = {cluster_order[i]: i for i in range(len(cluster_order))}
final_df['Cluster'] = final_df['Cluster'].map(cluster_mapping)

# Plotting
plt.figure(figsize=(10, 6))
sns.scatterplot(x=final_df.index, y=final_df['Final Marks'], hue=final_df['Cluster'], palette='viridis', s=200)
plt.xlabel("Student Index")
plt.ylabel("Final Marks")
plt.title("K-Means Clustering of Final Marks")
plt.legend(title="Cluster", loc="center left", bbox_to_anchor=(1, 0.5))
plt.show()
```
